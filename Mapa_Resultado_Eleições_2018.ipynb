Agora que já plotamos o mapa, vamos carregar os dados da votação por municípo

```python
df_votacao = pd.read_csv("/content/drive/My Drive/Colab Files/eleicoes/votacao_candidato_munzona_2018_BR.csv", sep=";", encoding="latin1")

df_votacao.head().T
```

Agora iremos separar somente a voração referente ao segundo turno

```python
df_votacao = df_votacao[df_votacao["NR_TURNO"] == 2] 
```

```python
df_votacao.sample(3).T
```

Agora precisamos ver quantos municípios únicos pesquisando pelo cógigo do município

```python
df_votacao["CD_MUNICIPIO"].nunique()
```

Existem municípios que também tiveram votação que estão em outros países

```python
df_votacao[df_votacao["SG_UF"] == "ZZ"].sample(4).T
```

Vamos verificar a quantidade de municípios fora do Brasil pelo Código do Municipio

```python
df_votacao[df_votacao["SG_UF"] == "ZZ"]["CD_MUNICIPIO"].nunique()
```

Como nosso mapa mostrará somente os municipios dentro do país, podemos eliminar os que estão fora do Brasil

```python
df_votacao = df_votacao[df_votacao["SG_UF"] != "ZZ"]
```

Haviam 5741 municipios cadastrados, vejamos quantos ficaram depois de retirar os que estão fora do país

```python
df_votacao["CD_MUNICIPIO"].nunique()
```

OS dados aqui das eleições estão separados por zona eleitoral, dessa forma um município pequeno pode ter apenas mas um municipio grande pode ter várias

Vejamos a quantidade de zonas no município de Goiânia

```python
df_votacao[df_votacao["NM_MUNICIPIO"] == "GOIÂNIA"]
```

Agora vamos eliminar algumas colunas que não serão relevantes carregando o dataset novamente somente com as colunas que iremos utilizar

```python
df_votacao = df_votacao[["SG_UF", "CD_MUNICIPIO", "NM_MUNICIPIO", "NR_CANDIDATO", "NM_URNA_CANDIDATO", "QT_VOTOS_NOMINAIS"]]
```

```python
df_votacao.head()
```

Agora vamos agregar os dados por municipio e candidato

```python
df_votacao.groupby(["SG_UF", "CD_MUNICIPIO", "NM_MUNICIPIO", "NR_CANDIDATO", "NM_URNA_CANDIDATO"]).sum()
```

Salvar o agrupamento no proprio dataframe resentando o index

```python
df_votacao = df_votacao.groupby(["SG_UF", "CD_MUNICIPIO", "NM_MUNICIPIO", "NR_CANDIDATO",
                    "NM_URNA_CANDIDATO"]).sum().reset_index()

df_votacao.sample(10)
```

Agora que o index foi resetado vamos pesquisar todos os registros separadamente

Como exemplo, veremos a quantidade de votos de cada candidato em Goiânia

```python
df_votacao[df_votacao["NM_MUNICIPIO"] == "GOIÂNIA"]
```

Para saber qual candidato foi o que mais teve votos em cada cidade, ordenaremos o dataframe pela quantidade de votos de forma decrecesnte e eliminaremos o segundo registro, deixando apenas o candidato que ficou em primeiro lugar

```python
df_votacao.sort_values(by="QT_VOTOS_NOMINAIS", ascending=False).head(10)
```

Agora vamos aplicar a ordenação no próprio dataframe

```python
df_votacao.sort_values(by="QT_VOTOS_NOMINAIS", ascending=False, inplace=True)
```

Agora que está ordenado vamos eliminar as linhas duplicadas, mantendo apenas a primeira linha

```python
df_votacao.drop_duplicates(subset="CD_MUNICIPIO", keep="first", inplace=True)

df_votacao.head(10)
```

Vamos verifcar em quantos municipios cada candidato venceu

```python
df_votacao["NM_URNA_CANDIDATO"].value_counts()
```

Agrupando a quantidade de votos por candidato

```python
df_votacao.groupby(["NM_URNA_CANDIDATO", "QT_VOTOS_NOMINAIS"]).sum()
```

Verificando a quantidade de votos nos municipios em Goiás

```python
df_votacao[df_votacao["SG_UF"] == "GO"].head(10)
```

Para separar o candidato que ganhou a eleição iremos criar uma nova coluna com essa informação, onde True=Bolsonaro e False=Haddad

```python
df_votacao["RESULTADO"] = df_votacao["NR_CANDIDATO"].apply(lambda x: x == 17)
```

```python
df_votacao.head(10)
```

Agora que o dataframe da quantidade de votos está pronto veremos novamente o dataframe do mapa

```python
df_mapa.head()
```

Agora precisamos juntar os dois dataframes, porém não podemos juntá-los pelo nome da cidade pois existem cidades com mesmo nome em diferentes estados.

Vejam como exemplo a cidade de nome Ouro Branco

```python
df_votacao[df_votacao["NM_MUNICIPIO"] == "OURO BRANCO"]
```

Para fazer essa junção precisaremos de outro dataset com as informações de equivalência entre as informações do IBGE e do TSE, encontramos essa tagela disponível no github do Estadão

```python
link = "https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv"

df_equivalencia = pd.read_csv(link)

df_equivalencia.head()
```

Agora é necessário preparar os dataframes, para isso vamos reindexar o dataframe de votação com o código do município do TSE, mas antes vamos fazer uma cópia do dataframe para deixar o original intacto

```python
df_votacao_novo = df_votacao.copy()
```

```python
df_votacao_novo.set_index("CD_MUNICIPIO", drop=False, inplace=True)
df_votacao_novo.head()
```

Agora vamos reindexar o dataframe de equivalênica também pelo código do TSE

```python
df_equi_novo = df_equivalencia.copy()
df_equi_novo.set_index("COD_TSE", drop=False, inplace=True)
df_equi_novo.head()
```

Agora precisamos verificar a quantidade de registros no dois dataframes

```python
df_votacao_novo.info()
```

```python
df_equi_novo.info()
```

A quantidade de registros batem então podemos fazer um merge entre os dois dataframes

```python
df_votacao_equivalencia = df_votacao_novo.join(df_equi_novo)
df_votacao_equivalencia.head()
```

Vejamos as informações do novo dataframe com as informações de votação e equivalência

```python
df_votacao_equivalencia.info()
```

Vejamos também as informações do dataframe do mapa

```python
df_mapa.info()
```

Transformando a coluna do código IBGE para string no novo dataframe

```python
df_votacao_equivalencia["GEOCOD_IBGE"] = df_votacao_equivalencia["GEOCOD_IBGE"].astype(str)
df_votacao_equivalencia.info()
```

Agora vamos unificar os indices, primeiro alterando o indice do dataframe do mapa e depois o indice do dataframe da votação/equivalência

```python
df_mapa.set_index("CD_GEOCMU", drop=False, inplace=True)

df_votacao_equivalencia.set_index("GEOCOD_IBGE", drop=False, inplace=True)
```

Para juntar os dois dataframes iremos criar um novo dataframe a partir dos indices que foram altrerados e agora são identicos

```python
df_mapa_novo = df_mapa.join(df_votacao_equivalencia)
df_mapa_novo.head().T
```

```python
df_mapa_novo.info()
```

A coluna "AJUSTE" possui uma quantidade muito grande de valores faltantes (NULL) o que torna irrelevante por isso deve ser excluída

```python
df_mapa_novo.drop("AJUSTE", axis=1, inplace=True)
```

```python
df_mapa_novo.info()
```

Precisamos saber quais as cidades que estão sobrando

```python
df_mapa_novo[df_mapa_novo["NR_CANDIDATO"].isnull()]
```

Já sabendo as cidades que estão sobrando basta eliminá-las

```python
df_mapa_novo.dropna(inplace=True)
```

```python
df_mapa_novo.info()
```

Pronto!

Todas as informações foram analisadas e os dataframes foram centralizados em um único dataframe para podermos desenhar o mapa.

Onde: False=red e True=blue



```python
from matplotlib.colors import ListedColormap
cmap = ListedColormap(["red", "blue"])

#desenhando o mapa
fig, ax = plt.subplots(1, figsize=(18, 18))
ax = df_mapa_novo.plot(column="RESULTADO", cmap=cmap, legend=True, linewidth=0.2, ax=ax, edgecolor="grey")

# escondendo os eixos
ax.set_axis_off()

# Titulo
ax.set_title("Resultado Eleição 2018 por municipio 2º Turno", fontdict={"fontsize": "25", "fontweight": "2"})

# Notas de rodapé
ax.annotate("Fonte: IBGE, 2015; TSE, 2019; Jornal o Estado de São Paulo, 2019",
            xy=(0.1, 0.08), xycoords="figure fraction", horizontalalignment="left",
            verticalalignment="top", fontsize=10, color="#555555")

# Legenda
leg = ax.get_legend()
leg.get_texts()[0].set_text("Fernando Hadad")
leg.get_texts()[1].set_text("Jair Bolsonaro")

plt.show()
```
