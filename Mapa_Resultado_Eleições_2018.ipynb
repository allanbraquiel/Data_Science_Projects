```python
# Arquivo com o resultado da eleição de 2018
df_votacao = pd.read_csv("/content/drive/My Drive/Colab Files/eleicoes/votacao_candidato_munzona_2018_BR.csv", 
                         sep=";", encoding="latin1")

df_votacao.head().T
```

```python
#salvar o groupby em um dataframe normal
df_votacao = df_votacao.groupby(["SG_UF", "CD_MUNICIPIO", "NM_MUNICIPIO", "NR_CANDIDATO",
                    "NM_URNA_CANDIDATO"]).sum().reset_index()

df_votacao.sample(10)
```

```python
# ordenar o dataframe
df_votacao.sort_values(by="QT_VOTOS_NOMINAIS", ascending=False, inplace=True)

#drop nas linhas suplicadas mantendo a primeira
df_votacao.drop_duplicates(subset="CD_MUNICIPIO", keep="first", inplace=True)

df_votacao.head(10)
```

```python
# criar uma coluna com o resultado da eleição 
# em cada municipio true=Bolsonaro false=Haddad

df_votacao["RESULTADO"] = df_votacao["NR_CANDIDATO"].apply(lambda x: x == 17)
```

```python
# porque nao podemos usar o nome do municipio para juntar os dois dataframes:
# existem varias cidades como mesmo nome
df_votacao[df_votacao["NM_MUNICIPIO"] == "OURO BRANCO"]
```

```python
# carregar a tabela de correspondencia entre ibge e tse direto do github do estadao
link = "https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv"

df_equivalencia = pd.read_csv(link)

df_equivalencia.head()
```

```python
df_equivalencia.head(10)
```

```python
# preparando os dataframes
# reindexar o dataframe de votacao pelo codigo do tse
# primeiro fazendo uma copia
df_vot_novo = df_votacao.copy()
```

```python
df_vot_novo.set_index("CD_MUNICIPIO", drop=False, inplace=True)
df_vot_novo.head()
```

```python
#reindexar o dataframe de equivalencia pelo codigo tse
df_equi_novo = df_equivalencia.copy()
df_equi_novo.set_index("COD_TSE", drop=False, inplace=True)
df_equi_novo.head()
```

```python
# Verificando a quantidade de registros nos dois dataframes
df_vot_novo.info()
```

```python
df_equi_novo.info()
```

```python
# Criando um novo dataframe com base no merge de votação e equivalencia
df_vot_equi = df_vot_novo.join(df_equi_novo)
df_vot_equi.head()
```

```python
# juntar o dataframe do geopandas com o votação/equivalencia
df_vot_equi.info()
```

```python
df_mapa.info()
```

```python
# transforamr o coluna de codigo ibge do dataframe votacao/equivalencia
df_vot_equi["GEOCOD_IBGE"] = df_vot_equi["GEOCOD_IBGE"].astype(str)
df_vot_equi.info()
```

```python
# unificando os indices
# mudando o indice do mapa
df_mapa.set_index("CD_GEOCMU", drop=False, inplace=True)
# mudando o indice do votacao/equivalencia
df_vot_equi.set_index("GEOCOD_IBGE", drop=False, inplace=True)
```

```python
# juntar os dataframes
df_mapa_novo = df_mapa.join(df_vot_equi)
df_mapa_novo.head().T
```

```python
df_mapa_novo.info()
```

```python
# Excluido a coluna AJUSTE por haver muitos valores NULL
df_mapa_novo.drop("AJUSTE", axis=1, inplace=True)
```

```python
df_mapa_novo.info()
```

```python
# Verificando quais as cidades estão sobrando
df_mapa_novo[df_mapa_novo["NR_CANDIDATO"].isnull()]
```

```python
# eliminar as duas informações que estao sobrando
df_mapa_novo.dropna(inplace=True)
```

```python
df_mapa_novo.info()
```

```python
#desenhar o mapa false=red true=green

from matplotlib.colors import ListedColormap
cmap = ListedColormap(["red", "blue"])

#desenhando o mapa
fig, ax = plt.subplots(1, figsize=(15, 15))

ax = df_mapa_novo.plot(column="RESULTADO", cmap=cmap, legend=True, linewidth=0.2, ax=ax, edgecolor="grey")

# esconder os eixos
ax.set_axis_off()

#Titulo 
ax.set_title("Resultado Eleição 2018 por municipio 2º Turno", fontdict={"fontsize": "25", "fontweight": "2"})

# Nota de rodape
ax.annotate("Fonte: IBGE, 2015; TSE, 2019; Jornal o Estado de São Paulo, 2019",
            xy=(0.1, 0.08), xycoords="figure fraction", horizontalalignment="left",
            verticalalignment="top", fontsize=10, color="#555555")

# Legenda
leg = ax.get_legend()
leg.get_texts()[0].set_text("Fernando Hadad")
leg.get_texts()[1].set_text("Jair Bolsonaro")

plt.show()
```
