# Criando um modelo

Para criar um modelo basta passar o nome do modelo com entrada. Esta função retorna uma tabela com as pontuações da validação cruzada e um objeto de modelo treinado. Neste exemplo usaremos o Adaboost ("ada")

```
adaboost = create_model("ada")
```

O objeto criado adaboost tem diversos parâmetros veja os exemplos abaixo

```
adaboost.base_estimator_
```

```
adaboost.feature_importances_
```

# Ajustando o modelo

A função tune_modelo ajusta automaticamente os hiperparâmetros do modelo usando uma pesquisa de grade aleatória em um espaço de pesquisa predefinido retornando uma tabela com as pontuações da validação cruzada e um objeto do modelo treinado.

```
tune_adaboost = tune_model(adaboost)
```

# Modelos Ensemble

A função ensemble_model é usada para agrupar os modelos treinados, só é preciso um parâmetro, ou seja, um objeto treinado. Retorna uma tabela com o resultado da validação cruzada e um objeto do modelo treinado.

```
# criando um modelo de Decision Tree
dt = create_model("dt")

# Agrupando o modelo treinado
dt_bagged = ensemble_model(dt)
```

# Plotando os resultados

A avaliação do desempenho do modelo treinado pode ser feita usando a função plor_model, onde é inserido o objeto do modelo treinado e o tipo de gráfico descrito como uma string dentro da função plot_model.

```
adaboost = create_model("ada")

# AUC (Curva ROC)
plot_model(adaboost, plot = "auc")
```

```
# Limite de decisão
plot_model(adaboost, plot= "boundary")
```

```
# Precision Recall Curve (Curva de precisão)
plot_model(adaboost, plot= "pr")
```

```
# Validation Curve (curva de validação)
plot_model(adaboost, plot= "vc")
```
